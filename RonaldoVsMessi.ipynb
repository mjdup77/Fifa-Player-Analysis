## Cristiano Ronaldo vs. Lionel Messi over Time

```python
import zipfile
import os
import pandas as pd

# Define the path to the uploaded zip file and the extraction directory
zip_path = '/Users/penneo/Downloads/fifa_2022_datasets.zip'
extraction_path = '/Users/penneo/Downloads/fifa_2022_datasets'

# Unzip the file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extraction_path)

# List the files in the extracted directory
extracted_files = os.listdir(extraction_path)
extracted_files
```

```python
import pandas as pd

# Base path for the extracted datasets
base_path = '/Users/penneo/Downloads/fifa_2022_datasets'

# List of CSV files for all players from FIFA 15 to FIFA 22
player_files = ['players_15.csv', 'players_16.csv', 'players_17.csv', 'players_18.csv', 
                'players_19.csv', 'players_20.csv', 'players_21.csv', 'players_22.csv']

# Load and concatenate the datasets, adding a 'FIFA_version' column
all_players_data = []

for file in player_files:
    # Load the data
    df = pd.read_csv(os.path.join(base_path, file))
    
    # Add a column to specify the FIFA version
    fifa_version = file.split('_')[1].split('.')[0]  # Extract the year from the filename
    df['FIFA_version'] = int(fifa_version)
    
    all_players_data.append(df)

# Concatenate all dataframes into one
all_players_df = pd.concat(all_players_data, ignore_index=True)

# Display the first few rows of the combined dataframe
all_players_df.head()
```

```python
# Calculate the percentage of missing values for each column
missing_percentage = all_players_df.isnull().mean() * 100

# Display columns with more than 50% missing values
columns_with_high_missing = missing_percentage[missing_percentage > 50].index
missing_percentage[missing_percentage > 50].sort_values(ascending=False)
```

```python
# Potential columns to drop (based on their nature or high missing values, excluding goalkeeping data)
potential_drop_columns = [
    'player_url', 'player_face_url', 'club_logo_url', 'club_flag_url', 
    'nation_logo_url', 'nation_flag_url', 'club_loaned_from', 
    'nation_team_id', 'nation_position', 'nation_jersey_number', 
    'player_tags'
]

potential_drop_columns
```

```python
# Remove the specified columns, excluding 'player_traits'
columns_to_drop = [col for col in potential_drop_columns if col != 'player_traits']
all_players_df_cleaned = all_players_df.drop(columns=columns_to_drop)

# Display the first few rows of the cleaned dataframe
all_players_df_cleaned.head()
```

## Attributions Analysis

```python
# Extract data for Messi and Ronaldo
messi_data = all_players_df_cleaned[all_players_df_cleaned['short_name'] == 'L. Messi']
ronaldo_data = all_players_df_cleaned[all_players_df_cleaned['short_name'] == 'Cristiano Ronaldo']

# Ensure the data is sorted by FIFA version for both players
messi_data = messi_data.sort_values(by='FIFA_version')
ronaldo_data = ronaldo_data.sort_values(by='FIFA_version')

messi_data[['short_name', 'FIFA_version', 'overall', 'potential']], ronaldo_data[['short_name', 'FIFA_version', 'overall', 'potential']]
```

```python
import pandas as pd
import matplotlib.pyplot as plt 
# Correct the attribute name in our list
attributes = ['overall']

# Initialize the plot
plt.figure(figsize=(15, 12))

# Plot each attribute
for idx, attribute in enumerate(attributes, 1):
    plt.subplot(3, 3, idx)
    plt.plot(messi_data['FIFA_version'], messi_data[attribute], marker='o', label='L. Messi', color='blue')
    plt.plot(ronaldo_data['FIFA_version'], ronaldo_data[attribute], marker='o', label='Cristiano Ronaldo', color='red')
    plt.title(attribute.capitalize())
    plt.xlabel('FIFA Version')
    plt.ylabel('Rating')
    plt.legend()
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Adjust layout
plt.tight_layout()
plt.show()
```

## Traits

```python
# Extract player traits for Messi and Ronaldo for 2015 through 2022
traits_messi_all_years = messi_data[['FIFA_version', 'player_traits']]
traits_ronaldo_all_years = ronaldo_data[['FIFA_version', 'player_traits']]

# Extract unique traits for each year
def extract_unique_traits(row):
    if pd.notna(row['player_traits']):
        return set(row['player_traits'].split(', '))
    return set()

traits_messi_all_years['unique_traits'] = traits_messi_all_years.apply(extract_unique_traits, axis=1)
traits_ronaldo_all_years['unique_traits'] = traits_ronaldo_all_years.apply(extract_unique_traits, axis=1)

traits_messi_all_years, traits_ronaldo_all_years
```

```python
import seaborn as sns

# Function to prepare data for heatmap
def prepare_data_for_heatmap(data):
    traits_years = {}
    for _, row in data.iterrows():
        for trait in row['unique_traits']:
            if trait not in traits_years:
                traits_years[trait] = {}
            traits_years[trait][row['FIFA_version']] = 1  # Presence of trait
    df = pd.DataFrame(traits_years).T.fillna(0)  # Convert dictionary to dataframe and fill NaN with 0
    return df

# Prepare data for Messi and Ronaldo
heatmap_data_messi = prepare_data_for_heatmap(traits_messi_all_years)
heatmap_data_ronaldo = prepare_data_for_heatmap(traits_ronaldo_all_years)

# Plotting the heatmaps
fig, ax = plt.subplots(2, 1, figsize=(14, 14))

sns.heatmap(heatmap_data_messi, cmap="Blues", cbar=False, ax=ax[0])
ax[0].set_title('Traits Evolution for Lionel Messi (FIFA 15-22)')

sns.heatmap(heatmap_data_ronaldo, cmap="Reds", cbar=False, ax=ax[1])
ax[1].set_title('Traits Evolution for Cristiano Ronaldo (FIFA 15-22)')

plt.tight_layout()
plt.show()
```

```python
import numpy as np 

# Grouping FIFA attributes into more detailed domains/categories
detailed_domains = {
    'Attacking': ['attacking_finishing', 'attacking_heading_accuracy', 'attacking_short_passing', 'attacking_volleys'],
    'Skills': ['skill_dribbling', 'skill_curve', 'skill_fk_accuracy', 'skill_ball_control'],
    'Movement': ['movement_acceleration', 'movement_sprint_speed', 'movement_agility', 'movement_reactions'],
    'Power': ['power_shot_power', 'power_stamina', 'power_strength', 'power_long_shots'],
    'Mentality': ['mentality_aggression', 'mentality_positioning', 'mentality_vision', 'mentality_composure'],
    'Passing': ['attacking_crossing', 'skill_long_passing', 'mentality_vision', 'attacking_short_passing']
}

# Calculate the average score for each detailed domain for Messi and Ronaldo
def calculate_detailed_domain_averages(player_data):
    averages = {}
    for domain, attributes in detailed_domains.items():
        averages[domain] = player_data[attributes].mean(axis=1).mean()  # Average over years and then average over attributes
    return averages

messi_detailed_averages = calculate_detailed_domain_averages(messi_data)
ronaldo_detailed_averages = calculate_detailed_domain_averages(ronaldo_data)

# Visualize using radar/spider charts
labels_detailed = list(detailed_domains.keys())
messi_detailed_stats = [messi_detailed_averages[domain] for domain in labels_detailed]
ronaldo_detailed_stats = [ronaldo_detailed_averages[domain] for domain in labels_detailed]

angles_detailed = np.linspace(0, 2 * np.pi, len(labels_detailed), endpoint=False)
# Close the plot
messi_detailed_stats = np.concatenate((messi_detailed_stats, [messi_detailed_stats[0]]))
ronaldo_detailed_stats = np.concatenate((ronaldo_detailed_stats, [ronaldo_detailed_stats[0]]))
angles_detailed = np.concatenate((angles_detailed, [angles_detailed[0]]))
labels_detailed.append(labels_detailed[0])  # Add the first label to the end to close the plot

plt.figure(figsize=(12, 12))
ax = plt.subplot(111, polar=True)
ax.plot(angles_detailed, messi_detailed_stats, 'o-', linewidth=2, label="L. Messi", color='blue')
ax.fill(angles_detailed, messi_detailed_stats, alpha=0.25, color='blue')
ax.plot(angles_detailed, ronaldo_detailed_stats, 'o-', linewidth=2, label="Cristiano Ronaldo", color='red')
ax.fill(angles_detailed, ronaldo_detailed_stats, alpha=0.25, color='red')
ax.set_thetagrids(angles_detailed * 180/np.pi, labels_detailed)
ax.set_title("Messi vs Ronaldo: Detailed FIFA Attributes (Average 2015-2022)")
ax.grid(True)
ax.legend()

plt.show()
```
